```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
def transform(img, label, out, n): 
    file = open(img, "rb") 
    output = open(out, "w") 
    label = open(label, "rb")

    file.read(16)
    label.read(8)
    im = []

    for j in range(n):
        imag = [ord(label.read(1))]
        for k in range(28 * 28):
            imag.append(ord( file.read(1)))
        im.append(imag)

    for i in im:
        output.write(",".join(str(pixel) for pixel in i) + "\n")
    file.close()
    output.close()
    label.close()
transform("/Users/jayavaniakkiraju/Downloads/train-images-idx3-ubyte", 
        "/Users/jayavaniakkiraju/Downloads/train-labels-idx1-ubyte", "mnist_data_train.csv", 60000) 
```

```python
mnist_data_train_csv = pd.read_csv("/Users/jayavaniakkiraju/Downloads/mnist_data_train.csv",header=None)
```

```python
mnist_data_train_data = mnist_data_train_csv.iloc[:,1:]
```

```python
mnist_data_train_label = mnist_data_train_csv.iloc[:,0]
```

```python
mnist_data_train_data /= 255. # Pixels range 
```

```python
plt.figure(figsize=(20,9))
for lab in range(10):
    Average_Lab = np.average(mnist_data_train_data[mnist_data_train_label==lab],0)
    plt.subplot(1, 10, lab+1)
    plt.imshow(Average_Lab.reshape(28,28),cmap ="binary_r") 
```

```python
plt.figure(figsize=(20,9))
for labe in range(10):
    std_Labe = np.std(mnist_data_train_data[mnist_data_train_label==labe],0)
    plt.subplot(1, 10, labe+1)
    plt.imshow(std_Labe.to_numpy().reshape(28,28),cmap='Greens') 
```
